```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import math
```

```python
img = Image.open('apel.jpg').convert('L')
F = np.array(img)
```

```python
neg = 255 - F
```

```python
c = 255 / np.log(1 + np.max(F))
log_img = c * np.log(1 + F)
log_img = np.clip(log_img, 0, 255).astype(np.uint8)
```

```python
plt.figure(figsize=(14,4))
plt.subplot(1,3,1); plt.imshow(F, cmap='gray'); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(neg, cmap='gray'); plt.title("Negative"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(log_img, cmap='gray'); plt.title("Log Transform"); plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(1,3,1); plt.hist(F.flatten(), bins=256); plt.title("Hist Original")
plt.subplot(1,3,2); plt.hist(neg.flatten(), bins=256); plt.title("Hist Negative")
plt.subplot(1,3,3); plt.hist(log_img.flatten(), bins=256); plt.title("Hist Log")
plt.show()
```
